```python
# importamos las librerias necesarias para la realización de la práctica
import re
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# introducimos la url desde donde haremos scraping
req = requests.get('https://datosmacro.expansion.com/otros/coronavirus')

# si el status code es 200, significa que la página se ha cargado de forma correcta
if req.status_code == 200:
    print('OK. Conexión establecida')
    content = req.content
    
html = BeautifulSoup(req.text, "html.parser")

# aquí iremos guardando los diferentes datos.
datos_pagina = {}

i=0 
```

```python
paginas = html.find('tbody', {'tr': ''})
    
for datos in paginas:
    
    pais = datos.find('td').find('a').getText().split()[0]
    
    activos = datos.find_all('td', {'class': 'numero'})[0].getText()
    
    inc_recuperados = datos.find_all('td', {'class': 'numero'})[1].getText()
    
    recuperados = datos.find_all('td', {'class': 'numero'})[2].getText()
    
    inc_fallecidos = datos.find_all('td', {'class': 'numero'})[3].getText()
    
    fallecidos = datos.find_all('td', {'class': 'numero'})[4].getText()
    
    inc_confirmados = datos.find_all('td', {'class': 'numero'})[5].getText()
    
    confirmados = datos.find_all('td', {'class': 'numero'})[6].getText()
    
    confirmados_100_mil_hab = datos.find_all('td', {'class': 'numero'})[7].getText()
    
    
    datos_pagina[i] = [pais, activos, inc_recuperados, recuperados, inc_fallecidos, fallecidos, inc_confirmados,
                      confirmados, confirmados_100_mil_hab]
    
    i = i+1
    
    print (pais, activos, inc_recuperados, recuperados, inc_fallecidos, fallecidos, inc_confirmados,
                      confirmados, confirmados_100_mil_hab)
    
```

```python
datos_df = pd.DataFrame.from_dict(datos_pagina, orient = 'index', columns = ['País', 'Activos', 'Inc Recuperados',
                                                                      'Recuperados', 'Inc. Fallecidos', 
                                                                      'Fallecidos', 'Inc. Confirmados',
                                                                      'Confirmados', 'Confirmados 100 mil hab.'])
```
